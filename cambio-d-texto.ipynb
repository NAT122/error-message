```python
def imprimir_texto(texto):
    try:
        impresora = win32print.OpenPrinter(win32print.GetDefaultPrinter())
        win32print.StartDocPrinter(impresora, 1, ("Texto aleatorio", None, "RAW"))
        win32print.StartPagePrinter(impresora)
        win32print.WritePrinter(impresora, texto.encode('utf-8'))
        win32print.EndPagePrinter(impresora)
        win32print.EndDocPrinter(impresora)
        win32print.ClosePrinter(impresora)
        return True
    except Exception as e:
        print(f"Error: {e}")
        return False

# Ciclo para imprimir 6 veces
for i in range(3):
    texto_para_imprimir = crear_contenido_impresion()
    print(f"Impresión {i+1}: {texto_para_imprimir.strip()}")  # Para ver en consola
    
    if imprimir_texto(texto_para_imprimir):
        print(f"¡Impresión {i+1} exitosa!")
    else:
        print(f"Error en la impresión {i+1}")
    print("-" * 30)
```
